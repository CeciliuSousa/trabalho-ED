```python
# alg_0 tipo: bubble sort - fonte https://panda.ime.usp.br/panda/static/pythonds_pt/05-OrdenacaoBusca/OBubbleSort.html

# base para pesquisa de complexidade: https://dev.to/williamspader/complexidade-de-algoritmos-big-o-59m5

# iniciando a função de nome 'algoritmo' e passando um parâmetro chamado 'arr'
# complexidade O(1)
def algoritmo(arr):

    # buscando o tamanho de 'arr' e guardando o valor na variável 'n'
    # complexidade O(1)
    n = len(arr)
    
    # começando um laço FOR que percorre até o penúltimo índice de 'arr', armazenando o índice temporariamente em 'i'
    # complexidade O(n)
    for i in range(n-1):

        # percorrendo o vetor até o penúltimo elemento não ordenado, pois a cada iteração o maior valor é movido para a posição correta no final, ordenando este vetor
        # complexidade O(n-i-1)
        for j in range(0, n-i-1):

            # comparando se o valor na posição à esquerda é maior que o valor à direita
            # complexidade O(1)
            if arr[j] > arr[j+1]:

                # caso o valor da esquerda seja maior, realiza a troca de posições
                # complexidade O(1)
                arr[j], arr[j+1] = arr[j+1], arr[j]

# EXEMPLO

# atribuindo valores ao vetor 'arr'
# complexidade O(1)
# arr = [64, 34, 25, 12, 22, 11, 90]

# vetor de 100 numeros para testes de velocidade
arr = [903, 678, 234, 17, 983, 127, 429, 569, 332, 450, 721, 860, 945, 199, 630, 103, 918, 214, 547, 315, 671, 984, 452, 608, 379, 25, 812, 463, 289, 790, 722, 118, 567, 731, 305, 49, 684, 935, 142, 869, 290, 355, 730, 491, 72, 915, 832, 411, 156, 597, 259, 349, 778, 417, 968, 724, 50, 586, 927, 309, 637, 702, 842, 654, 111, 436, 823, 156, 729, 48, 611, 304, 791, 567, 862, 112, 919, 187, 724, 493, 341, 226, 621, 712, 840, 297, 855, 469, 712, 588, 450, 384, 737, 191, 206, 273, 624, 502, 767, 874, 663, 892, 131, 797, 424, 652, 499]

# chamando a função e passando os valores atribuídos ao vetor 'arr'
# complexidade O(n^2)
algoritmo(arr)

# mostrando o texto 'Array ordenado:' na primeira linha
# complexidade O(1)
print("Array ordenado:")

# criando um laço FOR para percorrer o vetor ordenado
# complexidade O(n)
for i in range(len(arr)):
    
    # imprimindo o vetor ordenado
    # complexidade O(1) para cada iteração
    print("%d" % arr[i], end=" ")
```

```python
# alg_00 tipo: insertion sort - fonte: https://awari.com.br/aprenda-a-implementar-o-algoritmo-de-ordenacao-insertion-sort-em-python/

# base para pesquisa de complexidade: https://dev.to/williamspader/complexidade-de-algoritmos-big-o-59m5

# iniciando a função de nome 'algoritmo' e passando um parâmetro chamado 'arr'
def algoritmo(arr):

    # começando um laço FOR a partir do segundo elemento até o último de 'arr'
    # complexidade O(n)
    for i in range(1, len(arr)):

        # armazenando o valor do elemento atual na variável 'key'
        # complexidade O(1)
        key = arr[i]

        # inicializando a variável 'j' com o índice anterior ao de 'key'
        # complexidade O(1)
        j = i - 1

        # começando um laço WHILE que move os elementos maiores que 'key' uma posição à frente
        # complexidade no pior caso O(n), dependendo do número de elementos maiores que 'key'
        while j >= 0 and key < arr[j]:

            # movendo o elemento arr[j] uma posição à frente
            # complexidade O(1)
            arr[j + 1] = arr[j]

            # decrementando 'j' para continuar verificando os elementos anteriores
            # complexidade O(1)
            j -= 1

        # inserindo 'key' na posição correta encontrada
        # complexidade O(1)
        arr[j + 1] = key

# EXEMPLO

# atribuindo valores ao vetor 'arr'
# complexidade O(1)
arr = [64, 34, 25, 12, 22, 11, 90]

# vetor de 100 numeros para testes de velocidade
# arr = [903, 678, 234, 17, 983, 127, 429, 569, 332, 450, 721, 860, 945, 199, 630, 103, 918, 214, 547, 315, 671, 984, 452, 608, 379, 25, 812, 463, 289, 790, 722, 118, 567, 731, 305, 49, 684, 935, 142, 869, 290, 355, 730, 491, 72, 915, 832, 411, 156, 597, 259, 349, 778, 417, 968, 724, 50, 586, 927, 309, 637, 702, 842, 654, 111, 436, 823, 156, 729, 48, 611, 304, 791, 567, 862, 112, 919, 187, 724, 493, 341, 226, 621, 712, 840, 297, 855, 469, 712, 588, 450, 384, 737, 191, 206, 273, 624, 502, 767, 874, 663, 892, 131, 797, 424, 652, 499]

# chamando a função e passando os valores atribuídos ao vetor 'arr'
# complexidade O(n^2)
algoritmo(arr)

# mostrando o texto 'Array ordenado:' na primeira linha
# complexidade O(1)
print("Array ordenado:")

# criando um laço FOR para percorrer o vetor ordenado
# complexidade O(n)
for i in range(len(arr)):
    
    # imprimindo o vetor ordenado
    # complexidade O(1) para cada iteração
    print("%d" % arr[i], end=" ")
```

```python
# alg_000 tipo: Selection sort - fonte: https://awari.com.br/selecao-de-ordenacao-em-python-aprenda-a-implementar-o-selection-sort/

# base para pesquisa de complexidade: https://dev.to/williamspader/complexidade-de-algoritmos-big-o-59m5

# iniciando a função de nome 'algoritmo' e passando um parâmetro chamado 'arr'
def algoritmo(arr):

    # armazenando o tamanho de 'arr' na variável 'n'
    # complexidade O(1)
    n = len(arr)

    # começando um laço FOR que percorre o vetor até o penúltimo elemento
    # complexidade O(n)
    for i in range(n-1):

        # inicializando a variável 'min_idx' com o valor de 'i' para marcar o índice do menor elemento encontrado
        # complexidade O(1)
        min_idx = i

        # começando um laço FOR que percorre o vetor a partir do próximo elemento de 'i'
        # complexidade O(n-i)
        for j in range(i+1, n):

            # comparando se o elemento no índice 'j' é menor que o elemento no índice 'min_idx'
            # complexidade O(1)
            if arr[j] < arr[min_idx]:

                # atualizando 'min_idx' para o índice 'j' caso um elemento menor seja encontrado
                # complexidade O(1)
                min_idx = j

        # trocando o elemento no índice 'i' com o menor elemento encontrado
        # complexidade O(1)
        arr[i], arr[min_idx] = arr[min_idx], arr[i]

# EXEMPLO

# atribuindo valores ao vetor 'arr'
# complexidade O(1)
arr = [64, 34, 25, 12, 22, 11, 90]

# vetor de 100 numeros para testes de velocidade
# arr = [903, 678, 234, 17, 983, 127, 429, 569, 332, 450, 721, 860, 945, 199, 630, 103, 918, 214, 547, 315, 671, 984, 452, 608, 379, 25, 812, 463, 289, 790, 722, 118, 567, 731, 305, 49, 684, 935, 142, 869, 290, 355, 730, 491, 72, 915, 832, 411, 156, 597, 259, 349, 778, 417, 968, 724, 50, 586, 927, 309, 637, 702, 842, 654, 111, 436, 823, 156, 729, 48, 611, 304, 791, 567, 862, 112, 919, 187, 724, 493, 341, 226, 621, 712, 840, 297, 855, 469, 712, 588, 450, 384, 737, 191, 206, 273, 624, 502, 767, 874, 663, 892, 131, 797, 424, 652, 499]

# chamando a função e passando os valores atribuídos ao vetor 'arr'
# complexidade O(n^2)
algoritmo(arr)

# mostrando o texto 'Array ordenado:' na primeira linha
# complexidade O(1)
print("Array ordenado:")

# criando um laço FOR para percorrer o vetor ordenado
# complexidade O(n)
for i in range(len(arr)):
    
    # imprimindo o vetor ordenado
    # complexidade O(1) para cada iteração
    print("%d" % arr[i], end=" ")
```

```python
# alg_0000 tipo: quick sort - fonte: https://awari.com.br/python-aprenda-a-ordenar-listas-rapidamente-com-quicksort/

# base para pesquisa de complexidade: https://dev.to/williamspader/complexidade-de-algoritmos-big-o-59m5

# iniciando a função de nome 'algoritmo' e passando um parâmetro chamado 'arr'
def algoritmo(arr):

    # caso base: se o tamanho do array é menor que 2, retorna o array como está
    # complexidade O(1)
    if len(arr) < 2:
        return arr
    else:
        # escolhendo o primeiro elemento como o pivô
        # complexidade O(1)
        pivot = arr[0]

        # criando uma lista 'less' com elementos menores ou iguais ao pivô
        # complexidade O(n) onde n é o número de elementos no array
        less = [x for x in arr[1:] if x <= pivot]

        # criando uma lista 'greater' com elementos maiores que o pivô
        # complexidade O(n) onde n é o número de elementos no array
        greater = [x for x in arr[1:] if x > pivot]

        # aplicando recursivamente o algoritmo para as sublistas 'less' e 'greater', e concatenando o pivô entre elas
        # complexidade total O(n log n) no caso médio e O(n^2) no pior caso
        return algoritmo(less) + [pivot] + algoritmo(greater)

# EXEMPLO

# atribuindo valores ao vetor 'arr'
# complexidade O(1)
arr = [64, 34, 25, 12, 22, 11, 90]

# vetor de 100 numeros para testes de velocidade
# arr = [903, 678, 234, 17, 983, 127, 429, 569, 332, 450, 721, 860, 945, 199, 630, 103, 918, 214, 547, 315, 671, 984, 452, 608, 379, 25, 812, 463, 289, 790, 722, 118, 567, 731, 305, 49, 684, 935, 142, 869, 290, 355, 730, 491, 72, 915, 832, 411, 156, 597, 259, 349, 778, 417, 968, 724, 50, 586, 927, 309, 637, 702, 842, 654, 111, 436, 823, 156, 729, 48, 611, 304, 791, 567, 862, 112, 919, 187, 724, 493, 341, 226, 621, 712, 840, 297, 855, 469, 712, 588, 450, 384, 737, 191, 206, 273, 624, 502, 767, 874, 663, 892, 131, 797, 424, 652, 499]

# chamando a função e passando os valores atribuídos ao vetor 'arr', e armazenando o resultado em 'sorted_arr'
# complexidade O(n log n) no caso médio e O(n^2) no pior caso
sorted_arr = algoritmo(arr)

# mostrando o texto 'Array ordenado:' seguido do vetor ordenado
# complexidade O(n)
print("Array ordenado:")
print(sorted_arr)
```
